```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
N = 20
x = np.linspace(0, 2, N).reshape(-1, 1)
x = x.reshape(-1, 1)
y_true = 3*x+4
noise = np.random.normal(0,1, size=(N,1))
y = y_true + noise
```

```python
def solve_normal_equation(x, y):
    x = np.concatenate((np.ones((x.shape[0], 1)), x), axis=1)
    theta = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)
    return theta

theta = solve_normal_equation(x, y)




model = LinearRegression()
model.fit(x, y)

theta_0_sk = model.intercept_
theta_1_sk = model.coef_[0]



y_pred_ne = theta[0] + theta[1] * x
y_pred_sk = model.predict(x)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, label="Training Data", color="black")
plt.plot(x, y_pred_ne, label="Normal Equation", linewidth=2)
plt.plot(x, y_pred_sk, linestyle="--", label="Sklearn", linewidth=2)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()



print("\nParameter Comparison Table")
print("-" * 50)
print(f"{'Parameter':<15}{'True Value':<15}{'Normal Eq':<15}{'Sklearn'}")
print("-" * 50)
print(f"{'theta_0':<15}{4:<15}{theta[0][0]:<.4} {theta_0_sk[0]:.4}")
print(f"{'theta_1':<15}{3:<15}{theta[1][0]:<.4} {theta_1_sk[0]:.4}")
```
