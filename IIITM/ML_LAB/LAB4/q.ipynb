```python
x=df.drop('chd', axis=1)
# x
```

```python
y=df['chd']
# y
```

```python
x=pd.get_dummies(x, drop_first=True)
x= x.astype(int)
# x.head()
```

```python
x_sm= sm.add_constant(x)
```

```python
y=pd.get_dummies(y, drop_first=True)
y=y.astype(int)
# y
```

```python
model = sm.Logit(y,x_sm).fit()
# model.summary()
```

```python
significant_features = model.pvalues[model.pvalues < 0.05].index
significant_features = significant_features.drop('const')
```

```python
x_sig = x[significant_features]
x_sig_sm = sm.add_constant(x_sig)
model2 = sm.Logit(y, x_sig_sm).fit()
```

```python
coeffs = model2.params
# coeffs
```

```python
probs = model2.predict(x_sig_sm)
fpr, tpr, thresholds = roc_curve(y, probs)
```

```python
optimal_idx = np.argmax(tpr-fpr)
optimal_threshold = thresholds[optimal_idx]
```

```python
y_pred = (probs>= optimal_threshold).astype(int)
```

```python
cm = confusion_matrix(y, y_pred)
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)

print("Confusion Matrix:\n", cm)
print("Precision:", precision)
print("Recall:", recall)
```
